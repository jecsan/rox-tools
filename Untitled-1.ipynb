```python
from mss import mss
import cv2
from PIL import Image
import numpy as np
from time import time
import win32gui

import win32com.client as comclt

import win32com.client as comclt
from time import sleep

import random
```

```python
hwnd = win32gui.FindWindow(None, '(Stoic)')
x0, y0, x1, y1 = win32gui.GetWindowRect(hwnd)

windowWidth = x1 - x0 # width
windowHeight = y1 - y0 # height

windowPosX = (windowWidth/3)*2 + x0
windowPosY = windowHeight/2 + y0

print(windowWidth,"x", windowHeight)
print(x0," x", y0)
print(windowPosX," x", windowPosY)
showdebugScreen = False

```

```python
mon = {'top': int(windowPosY), 'left':int(windowPosX), 'width':int(windowWidth/3), 'height':int(windowHeight/2)}
cast_img = cv2.imread('cast.png',cv2.IMREAD_UNCHANGED)
reel_img = cv2.imread('reel.png',cv2.IMREAD_UNCHANGED)
threshold = 0.75
wsh= comclt.Dispatch("WScript.Shell")

# wsh.AppActivate("(Stoic)") # select another application

sct = mss()
fishing = False
idle = 0 #todo tra

while 1:
    begin_time = time()
    sct_img = sct.grab(mon)
    img = Image.frombytes('RGB', (sct_img.size.width, sct_img.size.height), sct_img.rgb)
    screen_img = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)
    
    if(showdebugScreen):
     cv2.imshow('Fishing', np.array(screen_img))

    if(fishing == False and random.random() < .70):
        result = cv2.matchTemplate(screen_img, cast_img, cv2.TM_CCOEFF_NORMED)

        min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(result)        
        w = cast_img.shape[1]
        h = cast_img.shape[0]

        if(showdebugScreen):
          cv2.imshow('Cast Result',result)

        if(max_val >= threshold and fishing == False):
            print("Start Fishing...")
            sleep(random.randrange(11,31,1)*.01 )
            wsh.AppActivate("(Stoic)")
            wsh.SendKeys("k")
            fishing = True
            sleep(3)

    if(fishing):
        idle+=1
        if(idle>150):
            idle = 0
            fishing = False
            break
        reelResult = cv2.matchTemplate(screen_img, reel_img, cv2.TM_CCOEFF_NORMED)
        
        min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(reelResult)
        
        w = reel_img.shape[1]
        h = reel_img.shape[0]

        if(showdebugScreen):
          cv2.imshow('Reel Result',reelResult)
        # cv2.rectangle(reelResult, max_loc, (max_loc[0]+ w, max_loc[1]+ h), (0,255,255),2)

        if(max_val >= threshold ):
            print("Stop Fishing...")
            wsh.AppActivate("(Stoic)")
            if( random.random() < .9):
                sleep(0.05)
            
            sleep(random.randrange(5,13,1)*.01 )
            wsh.SendKeys("k")
            sleep(1)
            print("Fishing...")
            idle = 0
            fishing = False
    
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break
    else:
     if (cv2.waitKey(25) & 0xFF == ord('w')):
         fishing = False
         
cv2.destroyAllWindows()
```
